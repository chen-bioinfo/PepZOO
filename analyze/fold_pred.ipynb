```python
import torch
import os
import esm
import biotite.structure.io as bsio
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
model = esm.pretrained.esmfold_v1()
model = model.eval().cuda()

# Optionally, uncomment to set a chunk size for axial attention. This can help reduce memory.
# Lower sizes will have lower memory requirements at the cost of increased speed.
# model.set_chunk_size(128)
seq_list = [
    # 'RGSHIKGRWKTRCHRHI',
    # 'RGSHIWSRWKTRCHRHI',
    # 'RGDHIKSRWKLRCHQH',
    # 'RGDHIKSRWKLRCCQH',
    # 'RGDCIKSRWKLRCCQ',
    # 'RGDCIKSRWKLRCCP',
    # 'RGDCIKSRWKLRVCP',
    # 'RGDCIKSRWKLYVCP',
    # 'RSDCIKSRWKLYVCP',
    # 'RGDCIKSRWKLQVCP',
    # 'RGDHIKSRWKLQVCP',
    # 'RGDCIKSRWKLQVC',
    # 'RGDVIKSRWKLQVC',
    # 'RGDVIKSRWKLQVCF',
    # 'RSDVTKSRWKLQVC',
    # 'RGDVTKSRWKLQVC',
    # 'RSDVTKSRWKLQVHP',
    # 'RSDVTKSRWKLQVCP',
    # 'RSDVTQSRWKLQVCP',
    # 'RSDVTKSRWKLQVCPG',
    'GVSGHGQHGVHG',
    'GVGGCMQSGVHG'
]
id = 1
for sequence in seq_list:
    print(sequence)
    with torch.no_grad():
        output = model.infer_pdb(sequence)


    with open(f"pdb/{id}_{sequence[0]}{sequence[-1]}-{len(sequence)}_opt.pdb", "w") as f:
        f.write(output)
    struct = bsio.load_structure(f"pdb/{id}_{sequence[0]}{sequence[-1]}-{len(sequence)}_opt.pdb", extra_fields=["b_factor"])
    id += 1
    print(struct.b_factor.mean())  # this will be the pLDDT
# 88.3
```
