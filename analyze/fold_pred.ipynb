```python
import numpy as np
aa_list = ['A', 'R', 'N', 'D', 'C', 'E', 'Q', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V']
def clean_seq_set(seq_list):
    # 去除序列中多余的字符
    new_seq_list = list()
    for i, seq in enumerate(seq_list):   # 去除前面的特殊标记 <|endoftext|> '\n'
        seq = seq.replace('\n', '').replace('<|endoftext|>', '')
        seq = seq.replace(' ', '').upper()
        flag = False
        if len(seq) == 0:
            continue
        for aa in seq:
            if aa not in aa_list:
                flag = True
                break
        if flag:
            continue
        else:
            new_seq_list.append(seq)
    return new_seq_list

seq_list = np.load('/geniusland/home/luojiawei1/project2023/prefix_tuning/generate/Allalpha/prefix_PREFIX_TUNING_CAUSAL_LM_E200_LR0.001_BS8_ML512.npy')
seq_list = clean_seq_set(seq_list)
print(seq_list)
```
