```python
import pandas as pd
import csv
import numpy as np
from Bio import SeqIO
```

# 准备数据集

```python
path_set = [
    'data/amps/Anti-Gram-_amps.fasta', 
    'data/amps/Anti-SARS-CoV-2_amps.fasta', 
    'data/amps/Antibacterial_amps.fasta',
    'data/amps/Anticancer_amps.fasta',
    'data/amps/Antifungal_amps.fasta',
    'data/amps/Antimicrobial_amps.fasta',
    'data/amps/Antiparasitic_amps.fasta',
    'data/amps/Antiviral_amps.fasta',
    'data/amps/general_amps.fasta',
    'data/amps/Insecticidal_amps.fasta',
    'data/amps/patent_amps.fasta',
    'data/amps/specific_amps.fasta'
]
alphabet = list('ACDEFGHIKLMNPQRSTVWY')
```

```python

for path in path_set:
    iterator = SeqIO.parse(path,'fasta')
    p, _ = path.split('.')
    write_file = open(f'{p}.csv', 'w')
    writer = csv.writer(write_file)
    writer.writerow(['sequence'])
    
    for record in iter(iterator):
        flag = 1
        if len(record.seq) <= 25:
            for c in record.seq:
                if c not in alphabet:
                    flag = 0
                    break
            if flag == 1:
                writer.writerow([record.seq])
    write_file.close()
```

```python
dic = {}
f = open('data/amps/amp_test.csv','w')
writer = csv.writer(f)
writer.writerow(['sequence'])
for path in path_set:
    data = pd.read_csv(path)['sequence'].tolist()
    for seq in data:
            dic[seq] = 0
for key in dic.keys():
    writer.writerow([key])
f.close()   
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成网格点
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)

# 定义一个径向对称的二维高斯函数
Z = np.exp(-0.2 * (X**2 + Y**2))

# 绘制等高线图
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')

# 添加颜色条
plt.colorbar(contour)

# 设置标题和标签
plt.title('Concentric Circles Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 美化图像使其类似于提供的图像
plt.axis('equal')  # 坐标轴比例相同
plt.tight_layout()  # 自动调整子图参数使之填充整个图像区域

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义一个简单的函数
def f(x, y):
    return np.sin(x) ** 2 + np.cos(y) ** 2

# 生成网格点
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 创建一个图形对象
fig, ax = plt.subplots()

# 绘制等高线图
contours = ax.contour(X, Y, Z, levels=20, cmap='viridis')
# 添加颜色条
plt.colorbar(contours)

# 添加等高线标签
ax.clabel(contours, inline=True, fontsize=8)

# 设置标题和标签
ax.set_title('Simple Contour Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# 显示图像
plt.show()
```

```python
dic = {}
w_f = open('data/train_all.csv','w')
writer = csv.writer(w_f)
writer.writerow(['sequence'])
for path in path_set:
    r_f = open(path, 'r')
    reader = csv.reader(r_f)
    next(reader)
    for row in reader:
        dic[row[1]] = 0
    r_f.close()
for key in dic.keys():
    writer.writerow([key])
w_f.close()

    
```
